```python
from tensorflow import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from skimage import color

# 화질 개선
from keras.layers import Average
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()
```

```python
# x_train = np.array([color.rgb2ydbdr(img) for img in x_train])
# x_test = np.array([color.rgb2ydbdr(img) for img in x_test])
```

```python
plt.imshow(x_train[0])
plt.show()
```

```python
# red 계열만... 뽑아 보면 어떰?
scaled_train = x_train / 255
scaled_test = x_test / 255

# 데이터 분할
scaled_train, scaled_val, y_train, y_val = train_test_split(
    scaled_train,
    y_train, 
    test_size=0.3,
    stratify=y_train,
    random_state=666
)
```

```python
# 맛집 레시피가보자
model1 = keras.Sequential()

# 입력층
model1.add(keras.Input(shape = (32, 32, 3)))

# 이미지 전처리 넣어보기


# 이미지를 늘려보는건.. 어떠니...
# model1.add(keras.layers.Resizing(50, 50))

# 무작위 선명도 추가~
# model1.add(keras.layers.RandomSharpness(factor=1.0, value_range=(0, 255)))

# 돌려 돌려 돌림판~~
model1.add(keras.layers.RandomFlip())
# model1.add(keras.layers.AutoContrast(value_range=(0, 255))) => 들어가자마자! 박살나는 성능!

# model1.add(keras.layers.RandomPerspective(
#     factor=1.0,
#     scale=0.4,
#     interpolation="bilinear",
#     fill_value=0.0,
# ))




# 합성곱층 추가
model1.add(keras.layers.Conv2D(32, kernel_size = 3, activation="relu", padding = "same"))
model1.add(keras.layers.Conv2D(32, kernel_size = 3, activation="relu", padding = "same"))
model1.add(keras.layers.MaxPooling2D(2))


# 두 번째 합성곱 - 풀링층 추가
model1.add(keras.layers.Conv2D(64, kernel_size = 3, activation="relu", padding= "same"))
model1.add(keras.layers.Conv2D(64, kernel_size = 3, activation="relu", padding= "same"))
model1.add(keras.layers.MaxPooling2D(2))

model1.add(keras.layers.Conv2D(128, kernel_size = 3, activation="relu", padding= "same"))
model1.add(keras.layers.Conv2D(128, kernel_size = 3, activation="relu", padding= "same"))
model1.add(keras.layers.MaxPooling2D(2))
# 분류신경망으로 넘겨 ~~~~~==================================================================
model1.add(keras.layers.Flatten())

# 과대적합 방지
model1.add(keras.layers.Dense(100, activation="relu"))
model1.add(keras.layers.Dropout(0.2))
model1.add(keras.layers.Dense(100, activation="relu"))

# 출력층 ( 이것도 범주가 10개니깐! )
model1.add(keras.layers.Dense(10, activation = "softmax"))
```

```python
model1.summary()
```

```python
es_cb = keras.callbacks.EarlyStopping(patience=10, restore_best_weights= True)

# 모델 설정하고 
model1.compile(optimizer="Adamax", loss = "sparse_categorical_crossentropy", metrics=["accuracy"])

# 적용해라
history1 = model1.fit(
    scaled_train, 
    y_train, 
    epochs=100, 
    validation_data=(scaled_val, y_val), 
    callbacks=[es_cb], 
    batch_size= 32
)
```

```python
model1.evaluate(scaled_test, y_test)
```

```python
plt.figure()

plt.plot(history1.history["loss"], label = "train_loss")
plt.plot(history1.history["val_loss"], label = "val_loss")

plt.plot(history1.history["accuracy"], label = "train_accuracy")
plt.plot(history1.history["val_accuracy"], label = "val_accuracy")

plt.legend()
plt.xlabel("epoch")
plt.show()
```

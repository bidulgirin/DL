```python
# pip install tensorflow 하셈
from tensorflow import keras # 구글
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import cross_validate, train_test_split
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()
```

```python
# 스케일링 건너뜀
```

```python
x_noScaled_train = x_train.reshape(-1, 28 * 28)
```

```python
noScaled_train, noScaled_val, y_train, y_val = train_test_split(
    x_noScaled_train,
    y_train, 
    test_size = 0.2,
    stratify = y_train, 
    random_state = 34
)
```

```python
print(noScaled_train.shape, y_train.shape)
```

# 입력층과 밀집층 입력

```python
inputs = keras.Input(shape = (784,))
dense = keras.layers.Dense(10, activation = "softmax")


# 신경망 모델을 생성
model = keras.Sequential()

# 신경망 모델에 밀집층을 추가
model.add(inputs)
model.add(dense)
```

```python
model.summary()
```

```python
# 모델을 학습시키기 위한 학습과정을 설정
model.compile(loss = "sparse_categorical_crossentropy", metrics = ["accuracy"])
```

```python
# 모델 훈련
model.fit(noScaled_train, y_train, epochs = 5)
```

- 스케일링 안한것보다 스케일링한것이 loss 가 더 적은것을 알수있음

- 

```python
# 검증데이터로 성능 평가
model.evaluate(noScaled_val, y_val)
```

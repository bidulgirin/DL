```python
# 이미지를 텐서로 변경
transform = transforms.Compose([transforms.ToTensor()])
```

```python
trainset = datasets.FashionMNIST(
    root = "./data/",
    train = True,
    download = True,
    transform = transform
)

testset = datasets.FashionMNIST(
    root = "./data/",
    train = False,
    download = True,
    transform = transform
)
```

```python
batch_size = 32
```

```python
# 데이터 공급 객체 선언
train_loader = data.DataLoader(
    dataset = trainset,
    batch_size = batch_size
)

test_loader = data.DataLoader(
    dataset = testset,
    batch_size = batch_size    
)
```

```python
#torch.Size([배치사이즈, 채널(색), 높이 너비]) 확인 하는 방법
trainset[0][0] # 독립변수 1개 (이미지 1개)
trainset[0][0].shape # 채널, 높이, 너비
# 케라스와 순서가 다르다
```

# 기본 설계

```python
# 기본 설계
class Net(nn.Module):
    def __init__(self):
        '''모델 구조 정의'''
        super().__init__()
        self.fc1 = nn.Linear(784, 100) # 784개의 데이터, 100개의 출력값이 나온다 밀집층을 fc1라는 이름으로 부르겠다
        self.fc2 = nn.Linear(100, 10) # 100개의 입력값이 들어가면 10개의 출력값이 나오겠다
    
    def forward(self, x):
        # 입력 데이터 x : x.shape = torch.Size([배치사이즈, 채널(색), 높이 너비])

        x = x.view(-1, 28 * 28) # fc1 에 넣기 위해서 1차원 변환
        # x.shape = torch.Size([배치사이즈, 784])
        
        x = F.relu(self.fc1(x)) # 활성화 함수 적용하면서~~~~~ 층에 넣기
        # x.shape = torch.Size([배치사이즈, 앞 레이어의 출력개수])

        x = self.fc2(x)
        
        # 예측값 리턴
        return x
```

# 모델 선언

```python
# 모델 선언
model = Net().to(DEVICE)
```

# 옵티마이저 선언

```python
# 옵티마이저 선언
optimizer = optim.SGD(model.parameters(), lr = 0.01) # 경사하강법 쓰겠다~~~ 모든 파라미터 넣었다
```

- 이런식으로 옵티마이저로 뒷부분만 학습할수는 있는데 지금은 전체 학습 기준으로 한다~~

```python
# 학습 함수 정의(fit 은 없다!!!!! 다 수동이다!!!!)
def train(model, train_loader, optimizer):
    # 모델을 학습 모드로 전환
    model.train() # 와 개쩐다!


    # 엄밀히 말하자면 (data, target) = (독립변수, 종속변수)
    for batch_idx, (data, target) in enumerate(train_loader): # 이 전체가 1 에포크임
        # 학습 데이터를 DEVICE로 보냄
        data = data.to(DEVICE)
        target = target.to(DEVICE)

        # 매 이터레이션 마다 기울기를 계산하기 위해 zero_grad() 호출
        optimizer.zero_grad() # 위치는 상관없다(가중치 수정하고 나서 해도됨)

        # 실제 모델의 예측값 받아오기
        output = model(data) # 순전파 forward 자동 발동!!!!!!!!
        
        # 정답 데이터와의 cross entropy loss 계산
        # loss 는 mini batch의 클래스의 오차 평균값
        loss = F.cross_entropy(output, target)

        # 기울기 계산
        loss.backward() # 오차 역전파~~~!!!!!!!기울기 계산됨

        # 가중치 수정
        optimizer.step()
```

```python
# 검증하는 함수
def evaluate(model, test_loader):
    # 모델을 평가 모드로 전환
    model.eval()

    # 필요한 변수 초기화
    # test과정에서의 loss = test_loss
    # 실제 모델의 예측이 정답과 맞은 횟수 = correct
    test_loss = 0
    correct = 0

    with torch.no_grad(): # 평가 시에는 기울기 계산안하니까 no_grad 명시
        # test_loader 에 대한 독립변수, 종속변수
        for data, target in test_loader:
            data = data.to(DEVICE)
            target = target.to(DEVICE)
            output = model(data)

            # 모든 오차 더하기 (cross_entropy: 기본값은 평균)
            test_loss += F.cross_entropy(output, target, reduction = "sum").item()


            # 가장 큰 값을 가진 클래스가 모델의 예측
            pred = output.max(1, keepdim = True)[1] 

            # 예측과 정답을 비교하여 일치하던 correct 에 1 더하기
            # max 인 index 를 뽑아내기 위한 코드
            # output.max(axis = 1 , keepdim 은 차원을 유지하겠다~~~ )[1]
            

            # 맞힌 것 개수 구하기
             # eq() : 값이 일치하면 1, 아니면 0
            correct += pred.eq(target.view_as(pred)).sum().item()
    # 평균 loss 값 구하기
    test_loss /= len(test_loader.dataset)

    # 정확도 계산
    test_accuracy = 100 * correct / len(test_loader.dataset)
    return test_loss, test_accuracy
```

- 부연 설명 no_grad 를 명시하는 이유

```python
for epoch in range(1, 6):
    train(model, train_loader, optimizer)
    test_loss, test_accuracy = evaluate(model, test_loader)

    print(f"[{epoch}] Test Loss : {test_loss:.4f}, Accuracy : {test_accuracy:.2f}%")
```

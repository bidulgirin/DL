# 전이 학습(transfer learning)

<img src="./image/transfer_learning.png">

- 기존꺼 잘 써보자~ 우어어어어어어어어어어어어어어어어어어어어어어어어어어ㅓ

- 일밙거으로 합성곱 신경망 기반의 딥러닝 모델을 제대로 훈련시키려면 많은 양의 데이터가 필요함

- 하지만 큰 데이터 셋을 확보하고 훈련시키는데에 많은 돈과 시간이 필요

- 이러한 문제를 해결하기 위해 전이 학습을 이용

- 전이학습(transfer learning)

    - 아주 큰 데이터셋 써서 훈련된 모델의 가중치를 가져와 우리가 해결하려는 과제에 맞게 보정해서 사용하는 것
    - 이 때 아주 큰 데이터셋을 사용하여 훈련된 모델을 사전 학습 모델(사전 훈련된 모델) 이라고 함
    - 전이 학습을 이용하면 비교적 **적은 수의 데이터**를 가지고도 원하는 과제를 해결할 수 있음

```python
import os
import time
import glob
import numpy as np

import torch
import torchvision
import torchvision.transforms as transforms
import torchvision.models as models # 여기에 사전학습모델이 다~~~~~ 있다
import torch.nn as nn
import torch.optim as optim

import matplotlib.pyplot as plt
```

```python
data_path = "./data/catanddog/train" 

transform = transforms.Compose(
    [
        transforms.Resize([256, 256]), # 모든 사진 사이즈가 다름. 같은사이즈로 통일
        transforms.RandomResizedCrop(224), # 랜덤으로 사진 크롭해주는 기능
        transforms.RandomHorizontalFlip(), # 50% 퍼 확률로 뒤집힘
        transforms.ToTensor(), # 텐서 타입으로 변경
    ]
)
```

```python
train_dataset = torchvision.datasets.ImageFolder(
        data_path,
        transform = transform
)
```

```python
train_loader = torch.utils.data.DataLoader(
    train_dataset,
    batch_size =32,
    num_workers= 0,
    shuffle = True
)
```

```python
print(len(train_dataset))
```

```python
# 학습에 사용될 이미지 출력
samples, labels = next(iter(train_loader)) # train_loader 에서 데이터를 꺼내오기
classes = {0:"cat", 1:"dog"}

fig = plt.figure(figsize = (20,24))
# pytorch (채널, 높이, 너비)
# matplotlib(높이, 너비, 채널)
for i in range(24):
    a = fig.add_subplot(4, 6, i +1)
    a.set_title(classes[labels[i].item()])
    a.axis("off")
    # pytorch 와 matplotlib 이 기대하는 차원이 다르기때문에 transpose
    a.imshow(np.transpose(samples[i].numpy(), (1,2,0)))

plt.subplots_adjust(bottom = 0.2, top=0.6, hspace=0.3 )
plt.show()
```

# 사전학습 모델 준비

```python
# 경량화 버전 resnet18 사용 하겠다

# 매개변수를 보면서 사용법을 익힌다
```

```python
resnet18 = models.resnet18(weights = "ResNet18_Weights.IMAGENET1K_V1")
```

## ResNet18

- ImageNet 데이터베이스의 100만개가 넘는 영상을 이용하여 훈련된 신경망

- 입력 제약이 매우 크고 충분한 **메모리(RAM)**가 없으면 학습속도가 느려지는 단점이 있음

```python
def set_parameter_requires_grad(model, feature_extracting = True):
    '''
    
    ResNet18이 파라키터 학습을 하지 않도록 고정

    '''

    if feature_extracting:
        for param in model.parameters():
            
            param.requires_grad = False # 역전파 중 파라미터들에 대한 변화를 계산할 필요 없음

            
```

```python
set_parameter_requires_grad(resnet18)
```

# 모델 학습

```python
def train_model(model, dataloaders, criterion, optimizer, device, num_epochs = 10, is_train = True):

    '''
    파라미터 : (모델, 학습데이터, 손실 함수, 옵티마이저, 장치)
    
    '''

    since = time.time() # 시간재려고
    
    acc_history = [] # 정확도
    loss_history = [] # 손실값
    best_acc = 0.0 # 가장좋은 정확도

    for epoch in range(num_epochs):
        print(f"Epoch {epoch}/ {num_epochs -1}") # 얼마나 돌고있니~ 
        print("-" * 10) # 절취선~


        running_loss = 0.0 # 해당 에포크의 로스 
        running_corrects = 0 # 해당 에포크의 정확도

        for inputs, labels in dataloaders:
            inputs = inputs.to(device) # 독립변수
            labels = labels.to(device) # 정속변수

            model.to(device) # 모델 장치에 보내주기
            optimizer.zero_grad() # 기울기 초기화
            outputs = model(inputs) # 순전파!!!!!!!!!!!!!!!!!!!!!!

            
            loss = criterion(outputs, labels) # 출력값, 정답값 비교 => 로스 나옴
            
            
            _, preds = torch.max(outputs, 1) # 예측 확률중에 axis = 1(열기준으로) 로 max 값과 max의 index 를 가져와라
            loss.backward() # 역전파~~~~~~~~~~
            optimizer.step() # 옵티마이저 돌면서 기울기 고쳐줌


            # 출력 결과와 레이블의 오차를 계산한 결과를 누적하여 저장
            running_loss += loss.item() * inputs.size(0) 
                
            # criterion(reduction = Sum) 을 하는것과 
            # running_loss +=  loss.item() * inputs.size(0) 하는것 같습니다~~
            
            running_corrects += torch.sum(preds == labels.data)


        epoch_loss = running_loss / len(dataloaders.dataset)

        # .double() 쓴 이유 타입 맞춰줄려고
        epoch_acc = running_corrects.double() / len(dataloaders.dataset)

        # 결과 출력~~~~~~~~~~~~~~~
        print(f"Loss : {epoch_loss:.4f} ACC : {epoch_acc:.4f}")

        # early stopping 거는 곳 ===================> 


        # 좋은 정확도 업데이트
        if epoch_acc > best_acc:
            best_acc = epoch_acc

        
        acc_history.append(epoch_acc.item())
        loss_history.append(epoch_loss)


        # 파일 저장
        torch.save(model.state_dict(), os.path.join("./data/catanddog/") + f"{epoch:02d}.pth")
        # .pth 파이토치 확장자명

        
    # 모든 에포크가 끝나고
    # 걸린 시간 알아보기~~~
    time_elapsed = time.time() - since
    print(f"Training complete in {time_elapsed // 60:.0f}m {time_elapsed % 60:.0f}s")
    print(f"Best Acc: {best_acc:4f}")
        
    return acc_history, loss_history
```

```python
# 파라미터 전달할것임
params_to_update = []

for name, param in resnet18.named_parameters():
    if param.requires_grad == True:
        params_to_update.append(param)
        print(name)
```

- 마지막 weight 와 bias 가 담긴다~

```python
optimizer = optim.Adam(params_to_update)
```

```python
optimizer
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
criterion = nn.CrossEntropyLoss()
```

```python
train_acc_hist, train_loss_hist = train_model(
    resnet18,
    train_loader, 
    criterion,
    optimizer,
    device
)
```

# 테스트 데이터 평가

```python
# 왜 경로를 못찾니.... 아니...
test_path = "./data/catanddog/test"
os.path.isdir(test_path)
```

```python
transform = transforms.Compose([
    transforms.Resize([224,224]),
    transforms.ToTensor()
])


test_dataset = torchvision.datasets.ImageFolder(
    root = test_path,
    transform = transform
)


test_loader = torch.utils.data.DataLoader(
    test_dataset,
    batch_size = 32,
    shuffle = True
)
```

```python
print(len(test_dataset))
```

```python
glob.glob("./data/catanddog/" + "*.pth") 
```

```python
# 가장 좋은 모델 파일 찾기!
def eval_model(model, dataloaders, device):
    since = time.time()
    acc_history = []
    best_acc = 0.0

    saved_models = glob.glob("./data/catanddog/" + "*.pth") # 경로에서 pth 확장자를 갖는 파일 가져오기

    saved_models.sort()
    
    print("saved_model", saved_models)

    for model_path in saved_models:
        print("Loading model", model_path)


        model.load_state_dict(torch.load(model_path)) # 저장되어있던 파라미터를 덮어씌운다
        model.eval() # 
        model.to(device)
        running_corrects = 0


        for inputs, labels in dataloaders:
            inputs = inputs.to(device)
            labels = labels.to(device)

            with torch.no_grad(): # 업데이트를 하지 않으니 자동 미분 계산이 필요 없음
                outputs = model(inputs)

            
            _, preds = torch.max(outputs.data, 1)
            preds[preds >= 0.5] = 1
            preds[preds < 0.5] = 0

            running_corrects += preds.cpu().eq(labels.cpu()).int().sum()

        epoch_acc = running_corrects.double() / len(dataloaders.dataset) # 모든 에포크에 대한 정확도

        print(f"ACC: {epoch_acc:.4f}")

        if epoch_acc > best_acc:
            best_acc = epoch_acc

        acc_history.append(epoch_acc.item())
        
        print()

    # 모든 검증 끝

    time_elapsed = time.time() - since

    print(f"Validation complete in {time_elapsed // 60 :.0f}m {time_elapsed % 60:.0f}s")

    print(f"Best Acc: {best_acc:.4f}")

    return acc_history
```

```python
# 함수 발동~~~
val_acc_hist = eval_model(resnet18, test_loader, device)
```

```python
plt.plot(train_acc_hist, label = "train_acc")
plt.plot(val_acc_hist, label = "val_acc")

plt.legend()
plt.show()
```

## 예측 이미지 출력

```python
def im_convert(tensor):
    # (채널, 높이, 너비) 변경해줘야함


    # 모델 연산에 영향을 줄 수 있음
    # 그걸 막으면서 진행
    image = tensor.clone().detach().numpy() # 기존 텐서의 내용을 복사한 텐서를 생성
    image = image.transpose(1, 2, 0) # 채널차원을 뒤로 옮기기
    image = image.clip(0, 1) # 값을 특정 범위로 제한 ( 0보다 큰애들 1로 맞춰주기)
    return image
```

- 텐서 복사 명령어 비교
- tensor.clone()
    - 메모리 : 새롭게 할당
    - 계산 그래프에 계속 상주 (기울기에 영향을 줌)

- tensor.detach()
    - 메모리 : 공유해서 사용
    - 계산 그래프에 상주하지 않음(기울기에 영향을 안줌)
    - **공유하고 있기 때문에 시각화 시 채널차원이 이상한데 라고 함~~~**
      
- tensor.clone().detach()
    - 메모리 : 새롭게 할당
    - 계산 그래프에 상주하지 않음(기울기에 영향을 안줌)

```python
classes = {
    0 : "cat",
    1 : "dog"
}

dataiter = iter(test_loader) # 이터레이터 상태로 변환 -> 데이터를 공급할수있는상태
images, labels = next(dataiter)
output = resnet18(images.to(device))

_, preds = torch.max(output, 1)
```

```python
fig = plt.figure(figsize = (24,4))


for idx in np.arange(20):
    ax = fig.add_subplot(2, 10, idx + 1, xticks = [], yticks = []) # 행, 열, 인덱, 축(삭제)
    plt.imshow(im_convert(images[idx]))
    ax.set_title(f"{str(classes[preds[idx].item()])}({str(classes[labels[idx].item()])})", color = ("green" if preds[idx] == labels[idx] else "red"))

#plt.subplots_adjust(bottom = 0.2, top = 0.6, hspace = 0)
plt.show()
```

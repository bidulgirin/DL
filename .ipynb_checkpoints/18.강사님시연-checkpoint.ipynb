```python
# 필요한것 import 
import pandas as pd
import numpy as np
# 형태소 분석기
from konlpy.tag import Okt
from tensorflow.keras.preprocessing.text import Tokenizer

from tensorflow import keras
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# 정규표현식
import re 
```

```python
train_df = pd.read_csv("./data/nsmc_ratings_train_pre.csv", index_col=False)
test_df = pd.read_csv("./data/nsmc_ratings_test_pre.csv", index_col=False)
```

```python
train_df.head()
```

# 리스트 형태로 변경하기

```python
train_df["token"] = train_df["token"].map(lambda x : eval(x))
test_df["token"] = test_df["token"].map(lambda x : eval(x))
```

```python
train_df.head()
```

## 독립변수 확인

```python
train_df.shape, test_df.shape
```

# 토큰화
- 하기전에 단어 수를 분석

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(train_df["token"])

total_cnt = len(tokenizer.word_index)
```

```python
# 등장 빈도가 2번 미만인 단어 수
rare_cnt = 0

for key, value in tokenizer.word_counts.items():
    if value < 2 :
        rare_cnt += 1
```

```python
rare_cnt
```

```python
rare_cnt / total_cnt
```

- 1번만 등장하는 단어는 전체 42.3%

## 전체에서 한번만 등장하는거 빼고 + 1(패딩)

```python
vocab_size = total_cnt - rare_cnt + 1
vocab_size
```

```python
# 토큰화
tokenizer = Tokenizer(num_words=vocab_size)
tokenizer.fit_on_texts(train_df["token"])

x_train = tokenizer.texts_to_sequences(train_df["token"])
x_test = tokenizer.texts_to_sequences(test_df["token"])
```

# 길이 분석(패딩)

```python
lengths = np.array([len(i) for i in x_train])
```

```python
plt.figure()

plt.hist(lengths)

plt.xlabel("length")
plt.ylabel("fre")

plt.show()
```

```python
np.mean(lengths), np.min(lengths), np.max(lengths), np.median(lengths)
```

```python
np.quantile(lengths, [0.1, 0.25, 0.5, 0.75, 0.9]) # 10퍼, 1사분위, 2사분위, 3사분위 ,90퍼
```

- 짧아서 20, 30 정도 고려할것임

# 그리스 아저씨

```python
train_seq = pad_sequences(x_train, maxlen = 30)
test_seq = pad_sequences(x_test, maxlen = 30)
```

```python
sub_seq, val_seq, y_sub, y_val = train_test_split(
    train_seq,
    train_df["label"], test_size = 0.2,
    stratify=train_df["label"], random_state=34
)
```

## 데이터는 shape 부터 찍어야합니다. 

- 나눈 데이터 확인

```python
sub_seq.shape,val_seq.shape,y_sub.shape,y_val.shape
```

- test 데이터 확인

```python
test_seq.shape
```

# 모델 설계

```python
model = keras.Sequential()
model.add(keras.Input(shape = (30,)))
model.add(keras.layers.Embedding(vocab_size, 128))

model.add(keras.layers.GRU(64, dropout = 0.5, return_sequences= True))
model.add(keras.layers.Dropout(0.5))
model.add(keras.layers.Dense(64))

model.add(keras.layers.GRU(128, dropout = 0.5, return_sequences= True))
model.add(keras.layers.Dropout(0.5))
model.add(keras.layers.Dense(128))

model.add(keras.layers.GRU(256, dropout = 0.5))
model.add(keras.layers.Dropout(0.5))
model.add(keras.layers.Dense(1, activation= "sigmoid"))
```

```python
model.summary()
```

- dense layer 로 정리 정리~~~~ 하는것이 특징

```python
model.compile(optimizer= "adam", loss = "binary_crossentropy", metrics = ["accuracy"])

es_cb = keras.callbacks.EarlyStopping(patience = 4, restore_best_weights=True)
```

```python
model.fit(
    sub_seq,
    y_sub,
    epochs =100,
    validation_data= (val_seq, y_val),
    batch_size=64,
    callbacks=[es_cb]
)
```

```python
model.evaluate(test_seq, test_df["label"])
```

# 예측하는 함수

```python
def sent_predict(text): 
    text = re.sub(r"[^ㄱ-ㅎ가-힣ㅏ-ㅣ]", " ", text)
    okt = Okt()
    text = okt.morphs(text, stem = True)

    encoded = tokenizer.texts_to_sequences([text])
    pad_new = pad_sequences(encoded, maxlen = 30)
    score = model.predict(pad_new)

    return score
```

```python
sent_predict("참 ~~~ 잘 한다")
```

```python
sent_predict("쥰내 귀여워요!")
```
